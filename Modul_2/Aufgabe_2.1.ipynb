```python
import pandas as pd
import requests
```

```python
url = "https://www.bibsonomy.org/json/search/sozialwissenschaft?items=1000"
url2 = "https://www.bibsonomy.org/json/search/statistik?items=1000"
result = requests.get(url)
result2 = requests.get(url2)
data = result.json()
data2 = result.json()
```

```python
df_bibsonomy = pd.DataFrame(data["items"])
df_bibsonomy2 = pd.DataFrame(data2["items"])
df_bibsonomy_gesamt = pd.concat([df_bibsonomy, df_bibsonomy2])
```

```python
df_bibsonomy_gesamt
```

```python
df_bibsonomy_selection = df_bibsonomy_gesamt[["label", "author", "year", "type"]]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection.type == "Publication"]
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection["author"].isnull() == False]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.drop_duplicates(subset = "label", keep = "first")
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection["year"].unique()
```

```python
df_bibsonomy_selection["year_correct"] = df_bibsonomy_selection.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.loc[df_bibsonomy_selection.year_correct == True]
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection["year"].unique()
```

```python
df_bibsonomy_selection = df_bibsonomy_selection.rename(columns={"label": "Titel", "author": "Autor", "year": "Jahr", "type": "Typ"})
```

```python
df_bibsonomy_selection
```

```python
df_bibsonomy_selection = df_bibsonomy_selection[["Titel", "Autor", "Jahr", "Typ"]]
```

```python
df_bibsonomy_selection
```

```python
import json
```

```python
df_bibsonomy_selection.to_json("output.json", orient="records")
```
